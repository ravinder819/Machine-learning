```python
num=9
prime=True
if num<=1:
    prime = False
else:
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            prime=False
            break
if prime:
    print(f"yes prime {num}")
else:
    print(f"no prime {num}")
```

```python
total = 0
while True:
    number = float(input("Enter a number (0 to stop): "))
    if number == 0:
        break
    total += number
print(f"The sum of all the numbers is {total}.")
```

```python
#sum of digit
num=12
sum=0
while num>0:
    digit=num%10
    sum=sum+digit
    num=num//10
print(f"sum of num is:{sum}")
```
